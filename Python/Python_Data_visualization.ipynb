```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')
iris
```

```python
# Create a scatter plot of sepal length vs. sepal width
sns.lmplot(x='sepal_length', y='sepal_width', data=iris,hue="species")
plt.title('Scatter Plot ')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# Create a scatter plot of sepal length vs. sepal width
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris, hue='species', palette='Set1', marker='*', s=250)

# Add a title
plt.title('Sepal Length vs. Sepal Width')

# Add labels to the axes
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# Add a legend
plt.legend(title='Species')

# Customize the gridlines
plt.grid(True, linestyle='--')

# Show plot
plt.show()
```

```python
# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a line plot of total bill vs. tip
sns.lineplot(x='total_bill', y='tip', data=tips)
plt.title('Line Plot of Total Bill vs. Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python

# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a line plot of total bill vs. tip
sns.lineplot(x='total_bill', y='tip', data=tips, color='blue', marker='o', markersize=8, linestyle='--')

# Add a title
plt.title('Line Plot of Total Bill vs. Tip')

# Add labels to the axes
plt.xlabel('Total Bill')
plt.ylabel('Tip')

# Customize the gridlines
plt.grid(True, linestyle='--')

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a bar plot of average tip amount by day
sns.barplot(x='day', y='tip', data=tips) # 1 cat , 1 nu
plt.title('Average Tip Amount by Day')
plt.xlabel('Day')
plt.ylabel('Average Tip Amount')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a count plot of the number of observations by day
sns.countplot(x='sex', data=tips)
plt.title('Number of Observations by Day')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a box plot of total bill by day
sns.boxplot(x='day', y='total_bill', data=tips) #1 cat, 1 nu
plt.title('Distribution of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a violin plot of total bill by day
sns.violinplot(x='day', y='total_bill', data=tips) #1 cat, 1 nu
plt.title('Distribution of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a strip plot of total bill by day
sns.stripplot(x='day', y='total_bill', data=tips,hue="sex") #1cat, 1 nu
plt.title('Strip Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a swarm plot of total bill by day
sns.swarmplot(x='day', y='total_bill', data=tips,hue="sex")
plt.title('Swarm Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a catplot of total bill by day
sns.catplot(x="time", y="total_bill",hue="smoker",col="day", data=tips, kind="bar") #kind= 'strip', 'swarm', 'box', 'boxen', 'violin', 'bar', 'count', and 'point'

# Add titles and labels
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
# Set the super title and adjust its vertical position
# Add a title and adjust its vertical position
plt.suptitle("Total Bill by Day, Separated by Time and Smoker Status", y=1.05)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create an lmplot of total bill vs. tip with hue based on smoker status
sns.lmplot(x="total_bill", y="tip", data=tips) #2 numerical

# Add titles and labels
plt.title("Linear Regression Plot of Total Bill vs. Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create an lmplot of total bill vs. tip with hue based on smoker status
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)

# Add titles and labels
plt.title("Linear Regression Plot of Total Bill vs. Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Show the plot
plt.show()
```
