# Messi vs Ronaldo comparison

```python
column=(fifa_data["Name"]=='L. Messi')|(fifa_data["Name"]=='Cristiano Ronaldo')
M_vs_R=fifa_data.loc[column]
column_to_drop=['ID','Photo','Flag','Club Logo','Special','Body Type','Real Face']
M_vs_R=M_vs_R.drop(columns=column_to_drop)
M_vs_R
```

```python
plt.title('Messi VS Ronaldo')
sns.barplot(x='Name', y='Overall',hue='Preferred Foot',data=M_vs_R)

plt.show()
```

```python
plt.title('Messi VS Ronaldo')
sns.barplot(x='Height', y='HeadingAccuracy',hue='Name',data=M_vs_R)
plt.show()
```
